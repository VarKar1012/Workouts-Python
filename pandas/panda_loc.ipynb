```python
# importing the module 
import pandas as pd 

# creating a sample dataframe 
data = pd.DataFrame({'Brand': ['Maruti', 'Hyundai', 'Tata', 
							'Mahindra', 'Maruti', 'Hyundai', 
							'Renault', 'Tata', 'Maruti'], 
					'Year': [2012, 2014, 2011, 2015, 2012, 
							2016, 2014, 2018, 2019], 
					'Kms Driven': [50000, 30000, 60000, 
									25000, 10000, 46000, 
									31000, 15000, 12000], 
					'City': ['Gurgaon', 'Delhi', 'Mumbai', 
							'Delhi', 'Mumbai', 'Delhi', 
							'Mumbai', 'Chennai', 'Ghaziabad'], 
					'Mileage': [28, 27, 25, 26, 28, 
								29, 24, 21, 24]}) 

# displaying the DataFrame 
display(data)
```

```python
print(data)
```

loc: access table based on condition

```python
data.info()
```

```python
data.loc[(data.Year>2012) & (data.Mileage>25)]
```

```python
data.loc[data.Brand == 'Maruti']
```

```python
data.loc[(data.City == 'Mumbai') & (data.Brand == 'Tata')]
```

loc[m:n]: slicing df : returns m to n rows (n included) - index starts from 0 onwards

```python
data.loc[2:5]
```

```python
data.loc[:4] #0-4
```

```python
data.loc[:] #all rows
```

```python
data.loc[::] #all rows
```

```python
data.loc[4:] # 4th to 8th rows
```

```python
data.loc[::3] #starts from 0th row with step as 3
```

```python
data.loc[6::] #6th row to 8th row
```

loc[index/condition, column]: changing value of the column

```python
# data.loc[['Mileage']]=20 #err
```

```python
data.loc[(data.Mileage < 25), ['Mileage']]=20
data
```

```python
data.loc[25, ['Mileage']]=20
data
```
